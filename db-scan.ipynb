```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN
from src.k_selectors import Density
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("train.csv")
target = df['tgt']
```

```python
def DBScanAnalyze(PCAs, target, epsilons, min_samples, metric='euclidean'):
    for eps in epsilons:
        for ms in min_samples:

            cluster = DBSCAN(eps=eps, min_samples = ms)
            label = cluster.fit_predict(PCAs)

            data = pd.DataFrame({
                'labels': label,
                'count': [1] * len(label),
                'classification': target
            })

            print(f"##### DBScan(eps = {eps}, min_samples = {ms})")
            print(data.groupby('labels').agg(
                {'count': 'count', 'classification': 'mean'}
            ))

            sns.scatterplot(x=PCAs[:,0], y=PCAs[:,1], hue=label, palette='bright', legend = False)
            plt.show()
```

```python
PCAs = pd.read_csv('count_PCA.csv').to_numpy()

DBScanAnalyze(
    PCAs, target, 
    epsilons = [1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9], 
    min_samples = [10]
)
```

```python
PCAs = pd.read_csv('first_PCA.csv').to_numpy()

DBScanAnalyze(
    PCAs, target, 
    epsilons = [0.02, 0.025, 0.03, 0.04], 
    min_samples = [3,5,10],
    metric = 'manhattan'
)
```
